## Desafio:

Com base nas aulas do módulo de Estatística, utilize os métodos da biblioteca Pandas para tratar o dicionário abaixo e:

- Trazer a média das vendas
- Criar um gráfico de barras vertical mostrando o mês de referência e o valor
- Criar um gráfico de linhas mostrando o mês de referência e o valor.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Dicionário de faturamento
dict_faturamento = {
    'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}
```

```python
df_faturamento = pd.DataFrame.from_dict(dict_faturamento)
```

```python
media_vendas = df_faturamento['valor'].mean()
```

```python
media_vendas
```

```python
# Converte a coluna 'data_ref' para o tipo datetime
df_faturamento['data_ref'] = pd.to_datetime(df_faturamento['data_ref'])

# Extrai o mês da coluna 'data_ref'
df_faturamento['mes'] = df_faturamento['data_ref'].dt.strftime('%m')
```

```python
# Cria gráficos de barras verticais
plt.figure(figsize=(12, 6))
plt.bar(df_faturamento['mes'], df_faturamento['valor'])

# Adiciona rótulo e títulos
plt.xlabel('Mês de referencia')
plt.ylabel('Valor')
plt.title('Gráfico de Barras Verticais - Mês de Referência vs. Valor')

# Rotaciona os rótulos do eixo x
plt.xticks(rotation=45, ha='right')

# Exibe o gráfico
plt.tight_layout()
plt.show()
```

```python
# Cria gráficos de barras verticais
plt.figure(figsize=(12, 6))
plt.plot(df_faturamento['mes'], df_faturamento['valor'], marker='o', linestyle='-', color='green')

# Adiciona rótulo e títulos
plt.xlabel('Mês de referencia')
plt.ylabel('Valor')
plt.title('Gráfico de Barras Verticais - Mês de Referência vs. Valor')

# Rotaciona os rótulos do eixo x
plt.xticks(rotation=45, ha='right')

# Exibe o gráfico
plt.tight_layout()
plt.show()
```
